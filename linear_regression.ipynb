```python
import yfinance as yf

univ = ['SPY', 'GOOG']

prices = yf.download(univ, period='1y')
prices
```

```python
adj_close = prices['Adj Close']
returns = adj_close / adj_close.shift() - 1
returns
```

```python
adj_close.plot()
```

```python
returns.plot()
```

```python
returns.mean()
```

```python
returns.std()
```

### Annualized Sharpe Ratio

```python
import math

returns.mean() * math.sqrt(252) / returns.std()
```
